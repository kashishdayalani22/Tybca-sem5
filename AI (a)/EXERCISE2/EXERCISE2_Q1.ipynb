```python
from nltk.chat.util import Chat, reflections
my_dummy_reflections={
    "mr":"doctor",
    "hello":"hey there",
    "play":"played"
}

pairs=[
    [
    r"my name is(.*)",
    ["hello %1,how are you today?",]
  ],
    [
        r"quit",
        ["bbye takecare. see you soon :)","it was nice talking to you. see you soon:)"]

    ],
]

def chatty():
   print("Hi, I'm Chatty and I chat alot ; type quit to leave")
   chat = Chat(pairs, my_dummy_reflections)
   chat.converse()
if __name__ == "__main__":

    chatty()


```
